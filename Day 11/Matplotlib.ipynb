```python
import matplotlib.pyplot as plt
```

## line plot

```python
x_axis=[1,2,3]
y_axis=[4,5,6]

plt.plot(x_axis,y_axis)
plt.show()
```

### color

```python
plt.plot(x_axis,y_axis,color='g')
```

```python
help(plt.plot)
```

```python
plt.plot(x_axis,y_axis,'bo')
```

```python
plt.title("Line Plot")
plt.plot(x_axis,y_axis,marker='x')
plt.show()
```

```python
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.plot(x_axis,y_axis,marker='x')
plt.show()
```

```python
x1_axis=[1,2,3]
y1_axis=[4,5,6]
x2_axis=[2,3,4]
y2_axis=[8,9,6]
plt.plot(x1_axis,y1_axis,'r')
plt.plot(x2_axis,y2_axis,'g')
plt.show()
```
